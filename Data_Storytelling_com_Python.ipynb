# Aula 021 - Data Storytelling com Python e Matplotlib

Data Storytelling significa conseguir comunicar sua mensagem por meio de visualizações gráficas, da melhor maneira possível e com o menor ruído possível na comunicação.

É uma skill desconhecida ou desprezada por muitos, mas que impacta diretamente em vários aspectos o seu projeto de Data Science e sua carreira como cientista de dados.

## 1. Exemplo de um gráfico sem técnicas de Data Storytelling

```
# importar as bibliotecas necessárias
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```
# dados a serem plotados
labels = ['Fortemente\nContra', 'Contra', 'Neutro', 'Favoravel', 'Fortemente\nFavoravel']
x = np.arange(len(labels))
width_bar = 0.4
antes = [22, 37, 17, 15, 9]
depois = [12, 10, 16, 33, 39] 
```

```
# criar uma figura para grafico de barras
fig, ax = plt.subplots(figsize=(8, 4))

# plotar antes e depois saparadamente
rects1 = ax.bar(x - width_bar/2, antes, width_bar, label='Antes')
rects2 = ax.bar(x + width_bar/2, depois, width_bar, label='Depois')

# labels do eixo
x = np.arange(len(labels))
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylabel('Porcentagem')

# título
ax.set_title('Apoio popular à Operação XYZ')

# legenda
ax.legend()

# plotar gráfico
fig.tight_layout()
plt.show()
```

## 2. Exemplo de um gráfico com técnicas de Data Storytelling

```
# configurar a visualização
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```
# configurando parâmetros das fontes
plt.rcParams['font.family'] = 'sans-serif'

# definindo cores do gráfico
CINZA1, CINZA2, CINZA3 = "#CFCDCE","#A2A1A3", "#787878"
AZUL2, AZUL1 = '#226089', '#46c3db' 
```

```
# ajustes para construir o gráfico


# definir cores por plot
cores_antes = [CINZA1, CINZA1, CINZA1, CINZA3, CINZA3]
cores_depois = [AZUL1, AZUL1, AZUL1, AZUL2, AZUL2]
```

```
# criar uma figura para grafico de barras
fig, ax = plt.subplots(figsize=(8, 4))

# plotar antes e depois saparadamente
rects1 = ax.bar(x - width_bar/2, antes, width_bar, label='Antes', color=cores_antes) #este color é para dar cor a cada label
rects2 = ax.bar(x + width_bar/2, depois, width_bar, label='Depois', color=cores_depois)# criamos uma lista com as cores e assume para cada label

#labels do eixo
x = np.arange(len(labels))
ax.set_xticks(x)
ax.set_xticklabels(labels)
# ax.set_ylabel('Porcentagem')

#remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
# remover eixo y
plt.gca().axes.get_yaxis().set_visible(False)
#remover xticks do eixo x
plt.tick_params(
    axis='both',
    which='both',
    bottom=False,
    top=False,
    labelbottom=True
)
#criar função para dar labels automaticamente. Escrever os valores nas barras
def autolabel(rects):
  for rect in rects:
    height = rect.get_height()
    ax.annotate(#serve para criar os números nas barras
        '{}%'.format(height),
        xy=(rect.get_x() + rect.get_width()/2, height),
        xytext=(0, -20), #offset para o texto
        textcoords='offset points',
        ha='center', #alinhamento centralizado
        va='bottom', #
        fontsize=12, # tamanho da fonte
        color='white', # cor do texto
        
    )
        
# plotar labels dentro das barras
autolabel(rects1)
autolabel(rects2)
# plotar título
plt.text(-0.6, 50,#Posição do texto largura e altura
         'Apoio Popular à Operação XYZ',
         fontsize=16,
         color=CINZA3,
         va='top')
# plotar subtitulo
plt.text(-0.6,46,
         'Efetividade da campanha\n'
         'junto à população ribeirinha'
         ,
         fontsize=12,
         color=CINZA2,
         va='top')
plt.text(4.5,39,
         '$\\bf{RESULTADO}$\n'
         'de uma campanha\n'
         'de marketing junto\n'
         'à população, onde o\n'
         'o apoio chegou a 39%'
         ,
         fontsize=12,
         color=AZUL2,
         va='top')

# plotar anotações do resultado da campanha

# plotar o gráfico
fig.tight_layout()
plt.show()
```
